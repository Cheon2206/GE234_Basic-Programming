## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


## 🔹 ตัวอย่างที่ 3: ใช้ GeoPandas โหลดและวิเคราะห์ข้อมูล Shapefile

```python
import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย
gdf = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/lab4/gpd_th_prov/tha_admbnda_adm1_rtsd_20190221.shp")

# แสดงข้อมูล 5 แถวแรก
print(gdf.head())
```


## 🔹 ตัวอย่างที่ 4: คำนวณพื้นที่ของแต่ละจังหวัด


```python
# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# สำหรับการคำนวณพื้นที่ที่แม่นยำ ควรแปลง CRS เป็น Projected CRS ก่อน
# ตัวอย่าง: ใช้ CRS ที่เป็นที่นิยมในประเทศไทย เช่น EPSG:32647 (UTM Zone 47N)
# gdf_projected = gdf.to_crs(epsg=32647)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางกิโลเมตร)
# หากใช้ gdf_projected: gdf_projected["area_sqkm"] = gdf_projected.geometry.area / 1e6
# หากยังคงใช้ gdf ที่เป็น Geographic CRS ค่าที่ได้อาจไม่แม่นยำตามคำเตือน
gdf["area_sqkm"] = gdf.geometry.area / 1e6

# แสดงพื้นที่จังหวัด 5 อันดับแรกที่ใหญ่ที่สุด
# แก้ไข PROV_NAME เป็น ADM1_TH หรือ ADM1_EN ตามคอลัมน์ที่มีอยู่จริง
print(gdf.nlargest(5, "area_sqkm")[["ADM1_TH", "area_sqkm"]])
```


## 🔹 ตัวอย่างที่ 5: ใช้ GeoPandas ทำ Spatial Join


```python
import os

# ตั้งค่า GDAL option เพื่อพยายามกู้คืนหรือสร้างไฟล์ .shx หากมีปัญหา
os.environ['SHAPE_RESTORE_SHX'] = 'YES'

# ตรวจสอบไฟล์ทั้งหมดในโฟลเดอร์ที่เก็บ Shapefile
folder_path = "/content/drive/MyDrive/Colab Notebooks/lab4/amp/"
print(f"ไฟล์ในโฟลเดอร์ {folder_path}:")
for file in os.listdir(folder_path):
    print(file)

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/lab4/amp/thailand_province_amphoe (1).shp")

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลที่เชื่อมโยงกัน
print(gdf_joined.head())
```

แบบฝึกหัด 1: ใช้ NumPy คำนวณค่า Mean, Max, Min ของค่า NDVI ในอาร์เรย์ที่สร้างขึ้นเอง

```python
print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.3f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.3f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.3f}")
```

แบบฝึกหัด 2: ใช้ GeoPandas โหลด Shapefile ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด

```python
# ตรวจสอบค่า CRS ของ GeoDataFrame จังหวัด
print(f"CRS ของข้อมูลจังหวัด: {gdf.crs}")

# คำแนะนำ: เพื่อการคำนวณพื้นที่ที่แม่นยำ ควรแปลง GeoDataFrame เป็น Projected CRS ก่อน
# ตัวอย่าง: แปลงเป็น EPSG:32647 (UTM Zone 47N สำหรับประเทศไทย)
# gdf_projected = gdf.to_crs(epsg=32647)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางกิโลเมตร)
# หากใช้ Geographic CRS (เช่น EPSG:4326) ค่าที่ได้อาจไม่แม่นยำ
gdf["area_sqkm"] = gdf.geometry.area / 1e6

# แสดงจังหวัดที่มีพื้นที่ 5 อันดับแรก (ใช้คอลัมน์ ADM1_TH สำหรับชื่อจังหวัด)
print("\nจังหวัดที่มีพื้นที่ 5 อันดับแรก (ตารางกิโลเมตร):")
print(gdf.nlargest(5, "area_sqkm")[["ADM1_TH", "area_sqkm"]])
```

แบบฝึกหัด 3: ใช้ GeoPandas ทำ Spatial Join ระหว่างข้อมูลจังหวัดและอำเภอ

```python
import os

# ตั้งค่า GDAL option เพื่อพยายามกู้คืนหรือสร้างไฟล์ .shx หากมีปัญหา
os.environ['SHAPE_RESTORE_SHX'] = 'YES'

# ตรวจสอบไฟล์ทั้งหมดในโฟลเดอร์ที่เก็บ Shapefile
folder_path = "/content/drive/MyDrive/Colab Notebooks/lab4/amp/"
print(f"ไฟล์ในโฟลเดอร์ {folder_path}:")
for file in os.listdir(folder_path):
    print(file)

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/lab4/amp/thailand_province_amphoe (1).shp")

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลที่เชื่อมโยงกัน
print(gdf_joined.head())
```

แบบฝึกหัด 4: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด

```python
import numpy as np

# เนื่องจากไม่มีข้อมูล NDVI จริงที่เชื่อมโยงกับ GeoDataFrame จังหวัด
# เราจะสร้างข้อมูล NDVI เฉลี่ยจำลองสำหรับแต่ละจังหวัดเพื่อแสดงวิธีการ
# สมมติว่ามีค่า NDVI เฉลี่ยสำหรับแต่ละจังหวัด
gdf['average_ndvi'] = np.random.uniform(low=0.1, high=0.9, size=len(gdf))

# หาจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด
province_highest_ndvi = gdf.loc[gdf['average_ndvi'].idxmax()]

print("จังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด:")
print(f"ชื่อจังหวัด: {province_highest_ndvi['ADM1_TH']}")
print(f"ค่า NDVI เฉลี่ย: {province_highest_ndvi['average_ndvi']:.3f}")
```
