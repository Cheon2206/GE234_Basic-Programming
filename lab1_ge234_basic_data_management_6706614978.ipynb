### **4.1 สร้างกราฟด้วย Matplotlib**  5 คะแนน
สร้างกราฟแสดงประชากรของแต่ละประเทศในภูมิภาค ASEAN โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python
# นำเข้าข้อมูล
from matplotlib import colors
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/drive/MyDrive/Colab Notebooks/asean_pop/aseans_pop_.csv'
geo_df = pd.read_csv(file_path)

# เปลี่ยนชื่อ
geo_data = geo_data.rename(columns={'name': 'Population 2026'})

geo_df['Population 2026'] = geo_df['Population 2026'].str.replace(',', '').astype(int)

# ตรวจสอบข้อมูล
print(geo_df.head())
print(geo_df.info())

# สร้างกราฟแสดงจำนวนประชากร
plt.figure(figsize=(15,20))
plt.bar(geo_df['Country'], geo_df['Population 2026'] / 1_000_000, color='#EF5579', zorder=2)
plt.title('ASEAN Population in 2026')
plt.xlabel('Country')
plt.ylabel('Population (Millions)')
plt.grid(True, axis='y', linestyle='--', alpha=0.7, zorder=1)
plt.grid(True, axis='x', linestyle='--', alpha=0.7, zorder=1)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```


### **4.2 การรวมข้อมูลเชิงพื้นที่และตาราง**
รวมข้อมูล Shapefile กับข้อมูลประชากรเพื่อแสดงข้อมูลประชากรในแต่ละจังหวัด โดยให้นักศึกษาทดลองหาข้อมูลเอง
    

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/apisit/thailand.json/master/thailand.json"
geo_data = gpd.read_file(url)

#Rename columns for clarity (e.g., ensuring we have a 'province_name' column)
geo_data = geo_data.rename(columns={'name': 'province_name'})

data = {
    'Province': ['Bangkok', 'Chiang Mai', 'Phuket', 'Khon Kaen', 'Chon Buri'],
    'Population (Million)': [5.5, 1.7, 0.4, 1.8, 1.5]
}
df = pd.DataFrame(data)

merged_data = geo_data.merge(df, left_on='province_name', right_on='Province', how='left')

fig, ax = plt.subplots(1, 1, figsize=(20, 30))

merged_data.plot(
    column='Population (Million)',
    cmap='RdBu',
    legend=True,
    legend_kwds={'label': "Population (Millions)"},
    missing_kwds={'color': 'lightgrey'},
    edgecolor='black',
    linewidth=0.5,
    zorder=2,
    ax=ax)

plt.title('Population by Province (Example Data)', fontsize=30)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.grid(True, axis='y', linestyle='--', alpha=0.7, zorder=1)
plt.grid(True, axis='x', linestyle='--', alpha=0.7, zorder=1)
plt.show()
```
