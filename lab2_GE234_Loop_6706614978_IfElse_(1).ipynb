## 🔹 ตัวอย่างที่ 1: การใช้ `for loop` เพื่อแสดงค่าพิกัด


```python

# พิกัดของสถานที่สำคัญในประเทศไทย
locations = {
    "Bangkok": (13.7563, 100.5018),
    "Chiang Mai": (18.7883, 98.9853),
    "Phuket": (7.8804, 98.3923),
    "Ubon Ratchathani": (15.2287, 104.8564)
}

# วนลูปแสดงค่าพิกัด
for city, coord in locations.items():
    print(f"{city}: Latitude {coord[0]}, Longitude {coord[1]}")
```


## 🔹 ตัวอย่างที่ 2: การใช้ `if-else` กรองข้อมูลพิกัด


```python

for city, coord in locations.items():
    if coord[0] > 16:
        print(f"{city} อยู่ทางภาคเหนือของประเทศไทย")
    else:
        print(f"{city} ไม่ใช่จังหวัดทางภาคเหนือ")
```


## 🔹 ตัวอย่างที่ 3: ใช้ `while loop` นับจำนวนจุดข้อมูล


```python

points = [(13.7563, 100.5018), (18.7883, 98.9853), (7.8804, 98.3923)]
count = 0

while count < len(points):
    print(f"Point {count + 1}: {points[count]}")
    count += 1
```


## 🔹 ตัวอย่างที่ 4: วิเคราะห์ข้อมูล Raster ด้วย `for loop`


```python

import numpy as np

# จำลองข้อมูล NDVI (Normalized Difference Vegetation Index)
ndvi = np.array([
    [0.2, 0.4, 0.5],
    [0.1, 0.3, 0.7],
    [0.6, 0.8, 0.2]
])

# วนลูปตรวจสอบค่าพืชพรรณ
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] > 0.3:
            print(f"พิกัด ({row}, {col}) มีพืชพรรณหนาแน่น (NDVI={ndvi[row, col]})")
```

แบบฝึกหัด 1: เขียน `for loop` เพื่อแสดงชื่อจังหวัดและพิกัดจากลิสต์ของคุณเอง

```python
# ชื่อจังหวัด
Province = ["Lampang, Chaiyaphum, Trat, Phatthalung, Uthai Thani, Nakhon Phanom, Ratchaburi"]

# พิกัดของจังหวัดตัวอย่าง 7 จังหวัดในประเทศไทย
locations = {"Lampang": (18.2888, 99.4908),
              "Chaiyaphum": (15.8068, 102.0310),
              "Trat": (12.2428, 102.5175),
              "Phatthalung": (7.6167, 100.0833),
              "Uthai Thani": (15.3794, 100.0245),
              "Nakhon Phanom": (17.3920, 104.7695),
              "Ratchaburi": (13.5367, 99.8171)}

# วนลูปแสดงค่าพิกัด
for Province, coord in locations.items():
    print(f"{Province}: \033[34m Latitude {coord[0]}\033[0m  , \033[35m Longitude {coord[1]}\033[0m")
# \033[...m = โค้ดพิเศษที่เรียกว่า ANSI escape codes เพื่อเปลี่ยนรูปแบบของข้อความ และต้องใส่ \033[0m ที่ท้ายข้อความเพื่อรีเซ็ตสีกลับเป็นปกติก่อนเริ่มรูปแบบใหม่ สำหรับรหัสสีจะเริ่มต้นที่ 30m-39m, 90m-97m , พื้นหลังเริ่มที่ 40m-47m , รูปแบบข้อความทั่วไป เช่น ตัวหนา ตัวบาง ตัวเอียง ฯลฯ เริ่มที่ 1m-9m
```

แบบฝึกหัด 2: ใช้ if-else เพื่อแยกแยะพื้นที่ที่เป็นภูเขาและพื้นที่ราบจากลิสต์ของพิกัด

```python
# แยกลักษณะภูมิประเทศในจังหวัดตัวอย่าง (สมมติเกณฑ์)

Province = ["Lampang, Chaiyaphum, Trat, Phatthalung, Uthai Thani, Nakhon Phanom, Ratchaburi"]

# พิกัดของจังหวัดตัวอย่าง 7 จังหวัดในประเทศไทย
locations = {"Lampang": (18.2888, 99.4908),
              "Chaiyaphum": (15.8068, 102.0310),
              "Trat": (12.2428, 102.5175),
              "Phatthalung": (7.6167, 100.0833),
              "Uthai Thani": (15.3794, 100.0245),
              "Nakhon Phanom": (17.3920, 104.7695),
              "Ratchaburi": (13.5367, 99.8171)}

for Province, coord in locations.items():
    if coord[0] > 15 :
        print(f"{Province} พื้นที่ภูเขา")
    else:
        print(f"{Province} พื้นที่ราบ")
```

แบบฝึกหัด 3: ใช้ while loop เพื่อวนลูปและแสดงค่าละติจูดจากลิสต์จนกว่าจะพบค่าต่ำกว่า 10°

```python
points = [(18.2888, 99.4908), (15.8068, 102.0310), (12.2428, 102.5175), (7.6167, 100.0833), (15.3794, 100.0245), (17.3920, 104.7695), (13.5367, 99.8171)]
count = 0

print("แสดงค่าละติจูดจนกว่าจะพบค่าต่ำกว่า 10°:")
while count < len(points):
    latitude = points[count][0]  # เพื่อดึงค่าละติจูดออกมา
    print(f"Point {count + 1}: Latitude {latitude}°")

    if latitude < 10:
        print(f"\033[91m***พบค่าละติจูดต่ำกว่า 10° ณ ละติจูดที่ {latitude}° The loop ends!!!")
        break  # หยุดลูปเมื่อพบค่าต่ำกว่า 10°

    count += 1
```

แบบฝึกหัด 4: ใช้ for loop วิเคราะห์ค่า NDVI และหาค่าพื้นที่ที่มีสภาพแห้งแล้ง (NDVI < 0.2)

```python
import numpy as np

# จำลองข้อมูล NDVI (Normalized Difference Vegetation Index)
ndvi = np.array([[0.2, 0.4, 0.5],
                 [0.1, 0.3, 0.7],
                 [0.6, 0.8, 0.2]])

print("พื้นที่ที่มีสภาพแห้งแล้ง (NDVI < 0.2):")
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] < 0.2:
            print(f"พิกัด ({row}, {col}) มีสภาพแห้งแล้ง (NDVI={ndvi[row, col]})\n")
        else:
            print(f"พิกัด ({row}, {col}) NDVI={ndvi[row, col]}")
```
