## 3.1 ‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÉ‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á (Indexing)

‡∏™‡∏ï‡∏£‡∏¥‡∏á‡πÉ‡∏ô Python ‡πÄ‡∏Å‡πá‡∏ö‡∏≠‡∏±‡∏Å‡∏Ç‡∏£‡∏∞‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏≥‡∏î‡∏±‡∏ö ‡πÄ‡∏£‡∏≤‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡πÑ‡∏î‡πâ‡∏î‡πâ‡∏ß‡∏¢ index (‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ó‡∏µ‡πà 0)

````python
# ‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£
text = "Python"
print("‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÅ‡∏£‡∏Å:", text[0])      # P
print("‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏ó‡∏µ‡πà 2:", text[1])     # y
print("‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢:", text[-1])  # n (‡∏ô‡∏±‡∏ö‡∏à‡∏≤‡∏Å‡∏ó‡πâ‡∏≤‡∏¢)
print("‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏ó‡∏µ‡πà 2:", text[-2])  # o

# ‡∏Å‡∏≤‡∏£‡∏ï‡∏±‡∏î‡∏™‡∏ï‡∏£‡∏¥‡∏á (Slicing)
print("‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 0-2:", text[0:3])      # Pyt
print("‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 2 ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πâ‡∏ô‡πÑ‡∏õ:", text[2:])  # thon
print("‡∏ñ‡∏∂‡∏á‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 4:", text[:4])      # Pyth
````

## 3.2 String Methods ‡∏™‡∏≥‡∏Ñ‡∏±‡∏ç

Python ‡∏°‡∏µ methods ‡∏°‡∏≤‡∏Å‡∏°‡∏≤‡∏¢‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏™‡∏ï‡∏£‡∏¥‡∏á

### 3.2.1 ‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå

````python
text = "Hello World"

print("‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î:", text.upper())       # HELLO WORLD
print("‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏•‡πá‡∏Å‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î:", text.lower())       # hello world
print("‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≥‡πÉ‡∏´‡∏ç‡πà:", text.title())     # Hello World
print("‡∏™‡∏•‡∏±‡∏ö‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå:", text.swapcase())           # hELLO wORLD
print("‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà:", text.capitalize())     # Hello world
````

### 3.2.2 ‡∏Å‡∏≤‡∏£‡∏ï‡∏±‡∏î‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á

````python
text = "   Hello World   "

print("‡∏ï‡∏±‡∏î‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏Ç‡πâ‡∏≤‡∏á:", text.strip())     # "Hello World"
print("‡∏ï‡∏±‡∏î‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡∏ã‡πâ‡∏≤‡∏¢:", text.lstrip())           # "Hello World   "
print("‡∏ï‡∏±‡∏î‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡∏Ç‡∏ß‡∏≤:", text.rstrip())           # "   Hello World"

# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏à‡∏£‡∏¥‡∏á: ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏õ‡πâ‡∏≠‡∏ô
user_input = "  Bangkok  "
clean_input = user_input.strip()
print(f"‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î: '{clean_input}'")
````

### 3.2.3 ‡∏Å‡∏≤‡∏£‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡πÅ‡∏•‡∏∞‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà

````python
text = "Python is easy. Python is powerful."

# ‡∏Ñ‡πâ‡∏ô‡∏´‡∏≤‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á
print("‡∏´‡∏≤ 'easy':", text.find("easy"))           # 10
print("‡∏´‡∏≤ 'difficult':", text.find("difficult")) # -1 (‡πÑ‡∏°‡πà‡πÄ‡∏à‡∏≠)

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ô‡∏µ‡πâ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
print("‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ 'Python' ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", "Python" in text)  # True
print("‡∏°‡∏µ‡∏Ñ‡∏≥‡∏ß‡πà‡∏≤ 'Java' ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", "Java" in text)      # False

# ‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà‡∏Ñ‡∏≥
new_text = text.replace("Python", "JavaScript")
print("‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà:", new_text)
````

### 3.2.4 ‡∏Å‡∏≤‡∏£‡πÅ‡∏¢‡∏Å‡πÅ‡∏•‡∏∞‡∏£‡∏ß‡∏°‡∏™‡∏ï‡∏£‡∏¥‡∏á

````python
# split() - ‡πÅ‡∏¢‡∏Å‡∏™‡∏ï‡∏£‡∏¥‡∏á‡πÄ‡∏õ‡πá‡∏ô list
sentence = "Python is easy to learn"
words = sentence.split()
print("‡πÅ‡∏¢‡∏Å‡∏Ñ‡∏≥:", words)

# ‡πÅ‡∏¢‡∏Å‡∏î‡πâ‡∏ß‡∏¢‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡πà‡∏ô‡∏ó‡∏µ‡πà‡∏Å‡∏≥‡∏´‡∏ô‡∏î
csv_data = "Bangkok,Thailand,10539000"
city_info = csv_data.split(",")
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏°‡∏∑‡∏≠‡∏á:", city_info)

# join() - ‡∏£‡∏ß‡∏° list ‡πÄ‡∏õ‡πá‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á
words = ["‡∏™‡∏ß‡∏±‡∏™‡∏î‡∏µ", "‡∏Ñ‡∏£‡∏±‡∏ö", "‡∏¢‡∏¥‡∏ô‡∏î‡∏µ‡∏ï‡πâ‡∏≠‡∏ô‡∏£‡∏±‡∏ö"]
greeting = " ".join(words)
print("‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏±‡∏Å‡∏ó‡∏≤‡∏¢:", greeting)

# ‡∏£‡∏ß‡∏°‡∏î‡πâ‡∏ß‡∏¢‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡πà‡∏ô‡∏≠‡∏∑‡πà‡∏ô
cities = ["Bangkok", "Chiang Mai", "Phuket"]
cities_str = ", ".join(cities)
print("‡πÄ‡∏°‡∏∑‡∏≠‡∏á‡∏ï‡πà‡∏≤‡∏á‡πÜ:", cities_str)
````

### 3.2.5 ‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏•‡∏±‡∏Å‡∏©‡∏ì‡∏∞‡∏Ç‡∏≠‡∏á‡∏™‡∏ï‡∏£‡∏¥‡∏á

````python
# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£/‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
text1 = "Python3"
text2 = "Python"
text3 = "12345"

print(f"'{text1}' ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", text1.isalnum())  # True
print(f"'{text2}' ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", text2.isalpha())    # True
print(f"'{text3}' ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", text3.isdigit())      # True

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå
print("'HELLO' ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÉ‡∏´‡∏ç‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", "HELLO".isupper())
print("'hello' ‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÄ‡∏•‡πá‡∏Å‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", "hello".islower())
````

## 3.3 ‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏™‡∏ï‡∏£‡∏¥‡∏á (String Formatting)

### 3.3.1 f-strings (‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥ - Python 3.6+)

````python
# f-string ‡∏û‡∏∑‡πâ‡∏ô‡∏ê‡∏≤‡∏ô
name = "‡∏™‡∏°‡∏ä‡∏≤‡∏¢"
age = 20
height = 175.5

# ‡πÅ‡∏ó‡∏£‡∏Å‡∏Ñ‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£
message = f"‡∏ä‡∏∑‡πà‡∏≠: {name}, ‡∏≠‡∏≤‡∏¢‡∏∏: {age} ‡∏õ‡∏µ, ‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á: {height} ‡∏ã‡∏°."
print(message)

# ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡πÉ‡∏ô‡∏™‡∏ï‡∏£‡∏¥‡∏á
print(f"‡∏≠‡∏µ‡∏Å 5 ‡∏õ‡∏µ {name} ‡∏à‡∏∞‡∏≠‡∏≤‡∏¢‡∏∏ {age + 5} ‡∏õ‡∏µ")

# ‡∏à‡∏±‡∏î‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°
pi = 3.14159265359
print(f"‡∏Ñ‡πà‡∏≤ pi = {pi:.2f}")          # 2 ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°
print(f"‡∏Ñ‡πà‡∏≤ pi = {pi:.4f}")          # 4 ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏°

# ‡∏à‡∏±‡∏î‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡∏°‡∏µ‡∏Ñ‡∏≠‡∏°‡∏°‡πà‡∏≤
population = 10539000
print(f"‡∏õ‡∏£‡∏∞‡∏ä‡∏≤‡∏Å‡∏£: {population:,} ‡∏Ñ‡∏ô")
````

### 3.3.2 .format() method

````python
# ‡πÉ‡∏ä‡πâ .format()
name = "‡∏™‡∏°‡∏´‡∏ç‡∏¥‡∏á"
gpa = 3.75

message = "‡∏ô‡∏±‡∏Å‡∏®‡∏∂‡∏Å‡∏©‡∏≤ {} ‡∏°‡∏µ GPA {:.2f}".format(name, gpa)
print(message)

# ‡πÉ‡∏ä‡πâ‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á
text = "{0} + {1} = {2}".format(5, 3, 5+3)
print(text)

# ‡πÉ‡∏ä‡πâ‡∏ä‡∏∑‡πà‡∏≠‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£
text = "‡πÄ‡∏°‡∏∑‡∏≠‡∏á: {city}, ‡∏õ‡∏£‡∏∞‡πÄ‡∏ó‡∏®: {country}".format(city="Bangkok", country="Thailand")
print(text)
````

## 3.4 ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ‡∏á‡∏≤‡∏ô‡∏à‡∏£‡∏¥‡∏á

### ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà 1: ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡πÅ‡∏•‡∏∞‡∏à‡∏±‡∏î‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ

````python
# ‡∏à‡∏≥‡∏•‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏õ‡πâ‡∏≠‡∏ô (‡∏°‡∏µ‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á‡πÄ‡∏Å‡∏¥‡∏ô‡πÅ‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡∏û‡∏¥‡∏°‡∏û‡πå‡πÑ‡∏°‡πà‡∏™‡∏°‡πà‡∏≥‡πÄ‡∏™‡∏°‡∏≠)
user_data = [
    "  somsak    ",
    "SOMYING  ",
    "  somchai"
]

# ‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
cleaned_names = []
for name in user_data:
    clean_name = name.strip().title()
    cleaned_names.append(clean_name)

print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏î‡∏¥‡∏°:", user_data)
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏∞‡∏≠‡∏≤‡∏î:", cleaned_names)
````

### ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà 2: ‡∏Å‡∏≤‡∏£‡πÅ‡∏¢‡∏Å‡∏û‡∏¥‡∏Å‡∏±‡∏î‡∏ó‡∏≤‡∏á‡∏†‡∏π‡∏°‡∏¥‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå

````python
# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏û‡∏¥‡∏Å‡∏±‡∏î‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö "latitude, longitude"
coordinate_str = "13.7563, 100.5018"

# ‡πÅ‡∏¢‡∏Å‡πÅ‡∏•‡∏∞‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç
lat_str, lon_str = coordinate_str.split(", ")
latitude = float(lat_str)
longitude = float(lon_str)

print(f"‡∏û‡∏¥‡∏Å‡∏±‡∏î: ‡∏•‡∏∞‡∏ï‡∏¥‡∏à‡∏π‡∏î {latitude}, ‡∏•‡∏≠‡∏á‡∏à‡∏¥‡∏à‡∏π‡∏î {longitude}")

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏û‡∏¥‡∏Å‡∏±‡∏î‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö WKT (Well-Known Text)
wkt_point = f"POINT({longitude} {latitude})"
print(f"WKT format: {wkt_point}")
````

### ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ó‡∏µ‡πà 3: ‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà

### ‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á List

````python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á list
fruits = ["apple", "banana", "orange"]
numbers = [1, 2, 3, 4, 5]
mixed = [1, "hello", 3.14, True]  # ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÄ‡∏Å‡πá‡∏ö‡∏´‡∏•‡∏≤‡∏¢‡∏ä‡∏ô‡∏¥‡∏î‡πÑ‡∏î‡πâ
empty_list = []  # list ‡∏ß‡πà‡∏≤‡∏á

print("‡∏ú‡∏•‡πÑ‡∏°‡πâ:", fruits)
print("‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç:", numbers)
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ú‡∏™‡∏°:", mixed)
print("‡∏ä‡∏ô‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•:", type(fruits))
````

### ‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô List

````python
fruits = ["apple", "banana", "orange", "mango", "grape"]

# ‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏î‡πâ‡∏ß‡∏¢ index (‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ó‡∏µ‡πà 0)
print("‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å:", fruits[0])      # apple
print("‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á:", fruits[1])    # banana
print("‡∏ï‡∏±‡∏ß‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢:", fruits[-1])  # grape (‡∏ô‡∏±‡∏ö‡∏à‡∏≤‡∏Å‡∏ó‡πâ‡∏≤‡∏¢)

# Slicing
print("3 ‡∏ï‡∏±‡∏ß‡πÅ‡∏£‡∏Å:", fruits[:3])   # ['apple', 'banana', 'orange']
print("‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà 2-4:", fruits[1:4]) # ['banana', 'orange', 'mango']
print("‡∏ó‡∏∏‡∏Å‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á:", fruits[::2])  # ['apple', 'orange', 'grape']
````

### ‡∏Å‡∏≤‡∏£‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô List

````python
fruits = ["apple", "banana", "orange"]
print("‡∏Å‡πà‡∏≠‡∏ô‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç:", fruits)

# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏Ñ‡πà‡∏≤
fruits[1] = "mango"
print("‡∏´‡∏•‡∏±‡∏á‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç:", fruits)

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
fruits.append("grape")        # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ó‡πâ‡∏≤‡∏¢‡∏™‡∏∏‡∏î
print("‡∏´‡∏•‡∏±‡∏á append:", fruits)

fruits.insert(1, "kiwi")      # ‡πÅ‡∏ó‡∏£‡∏Å‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏ó‡∏µ‡πà 1
print("‡∏´‡∏•‡∏±‡∏á insert:", fruits)

# ‡∏•‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
fruits.remove("orange")       # ‡∏•‡∏ö‡∏ï‡∏≤‡∏°‡∏Ñ‡πà‡∏≤
print("‡∏´‡∏•‡∏±‡∏á remove:", fruits)

deleted = fruits.pop()        # ‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢ ‡πÅ‡∏•‡∏∞‡∏™‡πà‡∏á‡∏Ñ‡πà‡∏≤‡∏Å‡∏•‡∏±‡∏ö
print("‡∏•‡∏ö‡∏ï‡∏±‡∏ß‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢:", deleted)
print("‡∏´‡∏•‡∏±‡∏á pop:", fruits)
````

### List Methods ‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏ö‡πà‡∏≠‡∏¢

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏¢‡∏≤‡∏ß‡∏Ç‡∏≠‡∏á list
print("‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å:", len(numbers))

# ‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏ï‡πà‡∏≤‡∏á‡πÜ
print("‡∏Ñ‡πà‡∏≤‡∏™‡∏π‡∏á‡∏™‡∏∏‡∏î:", max(numbers))
print("‡∏Ñ‡πà‡∏≤‡∏ï‡πà‡∏≥‡∏™‡∏∏‡∏î:", min(numbers))
print("‡∏ú‡∏•‡∏£‡∏ß‡∏°:", sum(numbers))

# ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡∏õ‡∏£‡∏≤‡∏Å‡∏è
print("1 ‡∏õ‡∏£‡∏≤‡∏Å‡∏è:", numbers.count(1), "‡∏Ñ‡∏£‡∏±‡πâ‡∏á")

# ‡∏´‡∏≤‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á
print("‡∏ï‡∏≥‡πÅ‡∏´‡∏ô‡πà‡∏á‡∏Ç‡∏≠‡∏á 5:", numbers.index(5))

# ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö
numbers.sort()  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏à‡∏≤‡∏Å‡∏ô‡πâ‡∏≠‡∏¢‡πÑ‡∏õ‡∏°‡∏≤‡∏Å
print("‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡πÅ‡∏•‡πâ‡∏ß:", numbers)

numbers.sort(reverse=True)  # ‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏à‡∏≤‡∏Å‡∏°‡∏≤‡∏Å‡πÑ‡∏õ‡∏ô‡πâ‡∏≠‡∏¢
print("‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏¢‡πâ‡∏≠‡∏ô‡∏Å‡∏•‡∏±‡∏ö:", numbers)

# ‡∏Å‡∏•‡∏±‡∏ö‡∏•‡∏≥‡∏î‡∏±‡∏ö
numbers.reverse()
print("‡∏Å‡∏•‡∏±‡∏ö‡∏•‡∏≥‡∏î‡∏±‡∏ö:", numbers)
````

````python
# ‡πÄ‡∏Å‡πá‡∏ö‡∏û‡∏¥‡∏Å‡∏±‡∏î‡πÄ‡∏°‡∏∑‡∏≠‡∏á‡∏ï‡πà‡∏≤‡∏á‡πÜ ‡πÉ‡∏ô‡∏õ‡∏£‡∏∞‡πÄ‡∏ó‡∏®‡πÑ‡∏ó‡∏¢
thailand_cities = [
    ("Bangkok", 13.7563, 100.5018),
    ("Chiang Mai", 18.7883, 98.9853),
    ("Phuket", 7.8804, 98.3923),
    ("Pattaya", 12.9236, 100.8825)
]

print("‡πÄ‡∏°‡∏∑‡∏≠‡∏á‡πÉ‡∏ô‡∏õ‡∏£‡∏∞‡πÄ‡∏ó‡∏®‡πÑ‡∏ó‡∏¢:")
for city in thailand_cities:
    name, lat, lon = city
    print(f"  {name}: ({lat}, {lon})")

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏°‡∏∑‡∏≠‡∏á‡πÉ‡∏´‡∏°‡πà
thailand_cities.append(("Khon Kaen", 16.4419, 102.8360))
print(f"\n‡∏°‡∏µ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {len(thailand_cities)} ‡πÄ‡∏°‡∏∑‡∏≠‡∏á")
````

## 4.2 Tuple (‡∏ó‡∏π‡πÄ‡∏û‡∏¥‡∏•)

## 4.3 Set (‡πÄ‡∏ã‡πá‡∏ï)

**Set** ‡πÄ‡∏õ‡πá‡∏ô‡πÇ‡∏Ñ‡∏£‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡πÄ‡∏Å‡πá‡∏ö‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà**‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô** ‡πÅ‡∏•‡∏∞**‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏•‡∏≥‡∏î‡∏±‡∏ö**

### ‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á Set

````python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á set ‡∏î‡πâ‡∏ß‡∏¢‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö‡∏õ‡∏µ‡∏Å‡∏Å‡∏≤ {}
fruits = {"apple", "banana", "orange"}
numbers = {1, 2, 3, 4, 5}

# ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏à‡∏≤‡∏Å list (‡∏à‡∏∞‡∏•‡∏ö‡∏Ñ‡πà‡∏≤‡∏ã‡πâ‡∏≥‡∏≠‡∏±‡∏ï‡πÇ‡∏ô‡∏°‡∏±‡∏ï‡∏¥)
numbers_with_duplicates = [1, 2, 2, 3, 3, 3, 4, 5]
unique_numbers = set(numbers_with_duplicates)

print("‡∏ú‡∏•‡πÑ‡∏°‡πâ:", fruits)
print("‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÄ‡∏î‡∏¥‡∏°:", numbers_with_duplicates)
print("‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥:", unique_numbers)
````

### ‡∏Å‡∏≤‡∏£‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡∏Å‡∏±‡∏ö Set

````python
regions = {"Asia", "Europe", "Africa"}

# ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å
regions.add("America")
print("‡πÄ‡∏û‡∏¥‡πà‡∏° America:", regions)

# ‡∏û‡∏¢‡∏≤‡∏¢‡∏≤‡∏°‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏ã‡πâ‡∏≥ (‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ú‡∏•)
regions.add("Asia")
print("‡πÄ‡∏û‡∏¥‡πà‡∏° Asia ‡∏ã‡πâ‡∏≥:", regions)  # ‡πÑ‡∏°‡πà‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡πÅ‡∏õ‡∏•‡∏á

# ‡∏•‡∏ö‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å
regions.remove("Africa")  # ‡∏•‡∏ö (‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏à‡∏∞ Error)
print("‡∏•‡∏ö Africa:", regions)

regions.discard("Antarctica")  # ‡∏•‡∏ö (‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Å‡πá‡πÑ‡∏°‡πà Error)

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏°‡∏µ‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
print("‡∏°‡∏µ Asia ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà:", "Asia" in regions)
print("‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å:", len(regions))
````

### Set Operations (‡∏Å‡∏≤‡∏£‡∏î‡∏≥‡πÄ‡∏ô‡∏¥‡∏ô‡∏Å‡∏≤‡∏£‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå)

````python
# ‡∏™‡∏≠‡∏á‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô
math_students = {"‡∏™‡∏°‡∏ä‡∏≤‡∏¢", "‡∏™‡∏°‡∏´‡∏ç‡∏¥‡∏á", "‡∏™‡∏°‡∏®‡∏£‡∏µ", "‡∏™‡∏°‡∏û‡∏á‡∏©‡πå"}
science_students = {"‡∏™‡∏°‡∏´‡∏ç‡∏¥‡∏á", "‡∏™‡∏°‡∏®‡∏£‡∏µ", "‡∏™‡∏°‡∏õ‡∏≠‡∏á", "‡∏™‡∏°‡∏´‡∏°‡∏≤‡∏¢"}

# Union (‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ô)
all_students = math_students | science_students
print("‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î:", all_students)

# Intersection (‡∏ï‡∏±‡∏î‡∏Å‡∏±‡∏ô - ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏ß‡∏¥‡∏ä‡∏≤)
both_subjects = math_students & science_students
print("‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏™‡∏≠‡∏á‡∏ß‡∏¥‡∏ä‡∏≤:", both_subjects)

# Difference (‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô - ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå)
only_math = math_students - science_students
print("‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå:", only_math)

# Symmetric Difference (‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏ß‡∏¥‡∏ä‡∏≤‡πÄ‡∏î‡∏µ‡∏¢‡∏ß)
one_subject_only = math_students ^ science_students
print("‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÄ‡∏û‡∏µ‡∏¢‡∏á‡∏ß‡∏¥‡∏ä‡∏≤‡πÄ‡∏î‡∏µ‡∏¢‡∏ß:", one_subject_only)
````

## 4.4 Dictionary (‡∏î‡∏¥‡∏Å‡∏ä‡∏±‡∏ô‡∏ô‡∏≤‡∏£‡∏µ)

**Dictionary** ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏π‡πà **key-value** ‡πÇ‡∏î‡∏¢ key ‡∏ï‡πâ‡∏≠‡∏á‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô ‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏Å‡∏≤‡∏£‡πÄ‡∏ä‡∏∑‡πà‡∏≠‡∏°‡πÇ‡∏¢‡∏á

### ‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á Dictionary

````python
# ‡∏™‡∏£‡πâ‡∏≤‡∏á dictionary ‡∏î‡πâ‡∏ß‡∏¢ {key: value}
student = {
    "name": "‡∏™‡∏°‡∏ä‡∏≤‡∏¢",
    "age": 20,
    "major": "Computer Science",
    "gpa": 3.75
}

print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏ô‡∏±‡∏Å‡πÄ‡∏£‡∏µ‡∏¢‡∏ô:", student)
print("‡∏ä‡∏ô‡∏¥‡∏î‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•:", type(student))

# Dictionary ‡∏ß‡πà‡∏≤‡∏á
empty_dict = {}
another_empty = dict()
````

### ‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô Dictionary

````python
student = {
    "name": "‡∏™‡∏°‡∏ä‡∏≤‡∏¢",
    "age": 20,
    "major": "Geography"
}

# ‡πÄ‡∏Ç‡πâ‡∏≤‡∏ñ‡∏∂‡∏á‡∏î‡πâ‡∏ß‡∏¢ key
print("‡∏ä‡∏∑‡πà‡∏≠:", student["name"])
print("‡∏≠‡∏≤‡∏¢‡∏∏:", student["age"])

# ‡πÉ‡∏ä‡πâ .get() (‡∏õ‡∏•‡∏≠‡∏î‡∏†‡∏±‡∏¢‡∏Å‡∏ß‡πà‡∏≤ ‡πÑ‡∏°‡πà Error ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ key)
print("‡∏™‡∏≤‡∏Ç‡∏≤:", student.get("major"))
print("GPA:", student.get("gpa", "‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•"))  # ‡∏ñ‡πâ‡∏≤‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÉ‡∏´‡πâ‡∏Ñ‡πà‡∏≤ default

# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡∏Ñ‡πà‡∏≤
student["age"] = 21
print("‡∏≠‡∏≤‡∏¢‡∏∏‡πÉ‡∏´‡∏°‡πà:", student["age"])

# ‡πÄ‡∏û‡∏¥‡πà‡∏° key-value ‡πÉ‡∏´‡∏°‡πà
student["gpa"] = 3.75
print("‡πÄ‡∏û‡∏¥‡πà‡∏° GPA:", student)
````

### Dictionary Methods ‡∏ó‡∏µ‡πà‡πÉ‡∏ä‡πâ‡∏ö‡πà‡∏≠‡∏¢

````python
city = {
    "name": "Bangkok",
    "country": "Thailand",
    "population": 10_539_000,
    "area_km2": 1568.7
}

# ‡∏î‡∏π keys ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
print("Keys:", city.keys())

# ‡∏î‡∏π values ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
print("Values:", city.values())

# ‡∏î‡∏π‡∏Ñ‡∏π‡πà key-value
print("Items:", city.items())

# ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
print("\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏°‡∏∑‡∏≠‡∏á:")
for key, value in city.items():
    print(f"  {key}: {value}")

# ‡∏•‡∏ö key
deleted_value = city.pop("area_km2")
print(f"\n‡∏•‡∏ö‡∏û‡∏∑‡πâ‡∏ô‡∏ó‡∏µ‡πà: {deleted_value}")
print("Dictionary ‡∏´‡∏•‡∏±‡∏á‡∏•‡∏ö:", city)
````

### ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Å‡∏≤‡∏£‡πÉ‡∏ä‡πâ Dictionary ‡πÄ‡∏Å‡πá‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏°‡∏∑‡∏≠‡∏á

````python
# ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏°‡∏∑‡∏≠‡∏á‡πÉ‡∏ô‡∏õ‡∏£‡∏∞‡πÄ‡∏ó‡∏®‡πÑ‡∏ó‡∏¢
thailand_cities = {
    "Bangkok": {
        "population": 10_539_000,
        "coordinates": (13.7563, 100.5018),
        "region": "Central"
    },
    "Chiang Mai": {
        "population": 1_735_000,
        "coordinates": (18.7883, 98.9853),
        "region": "North"
    },
    "Phuket": {
        "population": 416_582,
        "coordinates": (7.8804, 98.3923),
        "region": "South"
    }
}

# ‡πÅ‡∏™‡∏î‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
print("‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏°‡∏∑‡∏≠‡∏á‡πÉ‡∏ô‡∏õ‡∏£‡∏∞‡πÄ‡∏ó‡∏®‡πÑ‡∏ó‡∏¢:\n")
for city_name, city_data in thailand_cities.items():
    print(f"{city_name}:")
    print(f"  ‡∏õ‡∏£‡∏∞‡∏ä‡∏≤‡∏Å‡∏£: {city_data['population']:,} ‡∏Ñ‡∏ô")
    print(f"  ‡∏û‡∏¥‡∏Å‡∏±‡∏î: {city_data['coordinates']}")
    print(f"  ‡∏†‡∏π‡∏°‡∏¥‡∏†‡∏≤‡∏Ñ: {city_data['region']}")
    print()
````
